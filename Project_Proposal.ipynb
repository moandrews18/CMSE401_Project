Random number generation is a common computer programming tool. Random numbers are extremely important to such applications as statistic, game development, and cryptography. Many times solving these problems requires a vast amount of random numbers to be generated. In this project I will experiment with random number generation in parallel using NVIDIA's CUDA platform for parallel computing. For this I will also be using the Random123 library. Using these tools will give an idea to the speed and utility of running number generation in parallel. Finally, to prove the strength of parallel computing I will compare my solution to a non-parallelized solution in Python. A stretch goal for this project would be to use the randomly generated numbers to simulate a Monte Carlo problem.

### Part 1 - Software Exploration  
For this project the main software used will be NVIDIA's CUDA. This a parallel computing platform developed for high speed computation by giving users access to GPU-based accelerated processing. CUDA is not a programming language, rather, it is a tool to optimize programs within various languages. For this project I plan on using Python in Visual Studio to run my code.  
CUDA is not a tool I have used before so there will be a steep learning curve to getting started. Therefore I will be using these resources to get started:  
https://blogs.nvidia.com/blog/2012/09/10/what-is-cuda-2/
